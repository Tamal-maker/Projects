## DATA PREPROCESSING

```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)
```

```python
a = pd.read_csv('superstore.csv')
a.head(3)
```

```python
#Converting string to datetime and changing to desired format(SQL DATE FORMAT)

a['Order Date'] = pd.to_datetime(a['Order Date'], dayfirst=True).dt.strftime('%Y-%m-%d %H:%M:%S')     
```

```python
#Converting string to datetime and changing to desired format(SQL DATE FORMAT)

a['Ship Date'] = pd.to_datetime(a['Ship Date'], dayfirst=True).dt.strftime('%Y-%m-%d %H:%M:%S')
```
